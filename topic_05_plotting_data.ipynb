# Introduction to Plotting in Python with Matplotlib

Matplotlib is one of the most popular libraries for creating visualizations in Python.

In this notebook, you'll learn:
- How to make basic plots using lists
- How to customize your plots
- How to create different plot types (line, scatter, bar, histogram)
- How to make multiple plots in one figure


```python
# Importing the required libraries
import matplotlib.pyplot as plt
import math
import random
```

## Basic Line Plot

A line plot shows how one quantity changes with another.


```python
# Example: Plotting a sine wave
x = []
for i in range(0,63):
    x.append(i * 0.1)
    
y = []
for i in x:
    y.append(math.sin(i))

plt.plot(x, y)
plt.title("Sine Wave")
plt.xlabel("x (radians)")
plt.ylabel("sin(x)")
plt.grid(True)
plt.show()
```

## Customizing the Plot Appearance
You can change line styles, colors, add labels, and gridlines. See [here](https://matplotlib.org/stable/gallery/color/named_colors.html) for all the possible colours you can use in plots. See [here](https://matplotlib.org/stable/api/markers_api.html#module-matplotlib.markers) for all the types of markers you can use.


```python
x = [i * 0.1 for i in range(100)]
y1 = [math.exp(-0.1 * i) * math.sin(i) for i in x]
y2 = [math.exp(-0.1 * i) * math.cos(i) for i in x]

plt.figure(figsize=(8,5))
plt.plot(x, y1, 'b-')
plt.plot(x, y2, 'g:')

plt.title("Damped Oscillations")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.legend(['Damped Sine', 'Damped Cosine'])
plt.grid(True)
plt.show()
```

## Scatter Plots

Scatter plots show the relationship between two variables.


```python
# Example: Random data
random.seed(0)
x = [random.random() for _ in range(50)]
y = [random.random() for _ in range(50)]

plt.scatter(x, y, c='purple')
plt.title("Random Scatter Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

## Bar Charts

Bar charts are useful for comparing values between categories.


```python
categories = ['A', 'B', 'C', 'D']
values = [23, 45, 56, 78]

plt.bar(categories, values, color='skyblue')
plt.title("Category Comparison")
plt.xlabel("Category")
plt.ylabel("Value")
plt.show()
```
